### Create Graphs

```python
import glob
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_section_results(file):
    """
        requires tensorflow==1.12.0
    """
    X = []
    Y = []
    for e in tf.train.summary_iterator(file):
        for v in e.summary.value:
            if v.tag == 'Train_EnvstepsSoFar':
                X.append(v.simple_value)
            elif v.tag == 'Train_AverageReturn':
                Y.append(v.simple_value)
    return X, Y
```

```python
e = glob.glob("../../data/hw4_q4_reacher_horizon5*/events*")[0]
X, Y = get_section_results(e)
plt.plot(X[:len(Y)], Y, label="Horizon 5")

e = glob.glob("../../data/hw4_q4_reacher_horizon15*/events*")[0]
X, Y = get_section_results(e)
plt.plot(X[:len(Y)], Y, label="Horizon 15")

e = glob.glob("../../data/hw4_q4_reacher_horizon30*/events*")[0]
X, Y = get_section_results(e)
plt.plot(X[:len(Y)], Y, label="Horizon 30")

plt.legend()
plt.title("Comparing Horizon Lengths")
plt.xlabel("Timesteps")
plt.ylim((-600,-200))
plt.ylabel("Average Eval Reward")
info = "train_steps_per_iter: 1000 \nBatch Size: 800 \nn_iter: 15"
plt.figtext(.5,-.2, info, wrap=True, horizontalalignment="center", fontsize=12)
```
